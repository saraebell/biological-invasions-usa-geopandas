```python
#Analize 'text' features

def native_or_invasive_usa(row):
   if ("possibly native" in str(row.nativeregion)): 
    return('possibly-native')
   else: 
     return('not-native/invasive')
```

```python
native_or_invasive_usa_apply = df_species.apply(native_or_invasive_usa, axis='columns')
```

```python
df_species["native-invasive"] = native_or_invasive_usa_apply 
df_species.head(70)
```

```python
df_species['native-invasive'].value_counts()
```

```python
plt.pie(df_species['native-invasive'].value_counts(), autopct="%.1f%%",pctdistance=0.5,shadow=True,colors=['red','blue'],labels=['not-native/invasive','possibly-native'])
```
